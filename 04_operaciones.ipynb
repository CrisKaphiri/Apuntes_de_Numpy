```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
```

Operaciones aritméticas elemento a elemento

```python
print(f"La suma de los elementos es: {a + b}") # Suma
print(f"La resta de los elementos es: {a - b}") # Resta
print(f"La multiplicación de los elementos es: {a * b }") # Multiplicación
print(f"La división de los elementos es: {a / b}") # División
print(f"La potencia de los elementos es: {a ** b}") # Potencia
```

Operaciones matemáticas

```python
print(f"La raíz cuadrada de los elementos de la lista a: {np.sqrt(a) }")
print(f"e elevado a la potencia de los elementos de la lista a: {np.exp(a)}") 
print(f"El logaritmo natural de los elementos de a: {np.log(a)}")
print(f"el seno de los elementos de a: {np.sin(a)}")
```

Redondeo y manipulación de decimales

```python
c = np.array([1.63, 3.14, 5.35, 7.84])
print(c)
print(f"Redondea al entero más cercano: {np.round(c)}")
print(f"Redondea hacia abajo: {np.floor(c)}")
print(f"Redondea hacia arriba: {np.ceil(c)}")
print(f"Redondea con decimales especificos: {np.round(c, 1)}")
```
