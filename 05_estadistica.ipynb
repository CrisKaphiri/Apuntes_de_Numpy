```python
import numpy as np
```

Medias centrales

```python
arr = np.array([15, 20, 30, 40, 55])
print(f"Promedio: {np.mean(arr)}")
print(f"Mediana: {np.median(arr)}")
```

Valores extremos


```python
print(f"Mínimo: {np.min(arr)}")
print(f"Máximo: {np.max(arr)}")
```

Variabilidad

```python
print(f"Desviación estándar: {np.std(arr)}")
print(f"Varianza: {np.var(arr)}")
```

Suma y conteo

```python
print(f"Suma: {np.sum(arr)}")
print(f"Producto de elementos: {np.prod(arr)}")
print(f"Conteo: {np.count_nonzero(arr)}")
```
