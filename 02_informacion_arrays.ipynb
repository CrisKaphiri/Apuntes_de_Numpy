```python
import numpy as np
```

Dimensiones en los Arrays

```python
# Dimensión del array = 0
arr = np.array('10')
print(arr)
```

```python
# Dimensión del array = 1
arr = np.array([1, 2, 3, 4, 5])
arr
```

```python
# Dimensión del array = 2
arr = np.array([[10, 20, 30], [40, 50, 60]])
arr
```

```python
# Dimensión del array = 3
arr = np.array([[[1, 2, 3], [4, 5, 6]], [[10, 20, 30], [40, 50, 60]]])
arr
```

Tipo de dato de un array

```python
arr = np.array([1, 2, 3, 4])
arr.dtype
```

Numero de dimensiones

```python
a = np.array(42)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(f"El número de dimensiones del array es:" ,a.ndim)
print(f"El número de dimensiones del array es:" ,b.ndim)
print(f"El número de dimensiones del array es:" ,c.ndim)
print(f"El número de dimensiones del array es:" ,d.ndim)
```

Definir número de dimensiones

```python
arr = np.array([1, 2, 3, 4, 5], ndmin=5)
print(arr)
print('Número de dimensiones del array:', arr.ndim)
```
